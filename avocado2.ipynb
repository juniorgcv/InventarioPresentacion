```python
import statistics

def get_shape(data):
    return f"Filas: {data.shape[0]}\nColumnas: {data.shape[1]}"

def get_precio_stats(data, precio_col='AveragePrice'):
    return f"Precio mínimo: {data[precio_col].min()}\nPrecio máximo: {data[precio_col].max()}\nPrecio promedio {data[precio_col].mean()}"
```

```python
import seaborn as sns

def scatter_regiones(data, regiones=['Atlanta', 'Charlotte', 'Columbus']):
    data_regiones = data[data['region'].isin(regiones)]
    sns.scatterplot(data=data_regiones, x='year', y='AveragePrice', hue='region')
    plt.show()
```

```python
def scatter_regiones_vectorizado(data, regiones=['Atlanta', 'Charlotte', 'Columbus']):
    data_regiones = data[data['region'].isin(regiones)]
    plt.scatter(data_regiones['year'], data_regiones['AveragePrice'], c=data_regiones['region'])
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets

data = pd.read_csv('/content/avocado.csv')

# 1. Obtener el número de filas y columnas
print("Mostrando el numero de filas y columnas")
print(get_shape(data))
print()
print()

# 2. Mostrar los primeros 100 registros
print("Mostrando los primeros 100 registros")
print(data.head(100))
print()
print()

# 3. Mostrar los últimos 20 registros
print("Mostrando los ultimos 20 registros")
print(data.tail(20))
print()
print()

# 4. Precio mínimo, máximo y promedio del aguacate
print("Mostrando el precio maximo, minimo y promedio")
print(get_precio_stats(data))
print()
print()

# 5. Gráfico de dispersión con 3 regiones
print("Mostrando un grafico de dispercion")
scatter_regiones(data)
```
